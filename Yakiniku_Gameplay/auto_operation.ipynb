```python
# 比對所有圖片，取得顯示在桌面的圖片物件 (pillow / PIL Image)

btn_open_location = pyautogui.locateOnScreen('meat/game.png', confidence=0.9)

btn_open_point = pyautogui.center(btn_open_location)

pyautogui.doubleClick(btn_open_point)
sleep(0.5)

pyautogui.PAUSE = 0.01

btn_fullscreen_location = pyautogui.locateOnScreen('meat/fullscreen.png', confidence=0.9)

btn_fullscreen_point = pyautogui.center(btn_fullscreen_location)

pyautogui.click(btn_fullscreen_point)

sleep(0.5)
pyautogui.PAUSE = 0.01

btn_start_location = pyautogui.locateOnScreen('meat/start.png', confidence=0.9)

btn_start_point = pyautogui.center(btn_start_location)

pyautogui.click(btn_start_point)

meatLoc = []

# wireGrid = [
#     [880,385],
#     [880,470],
#     [1000,385],
#     [1000,470],
#     [1140,385],
#     [1140,470]
# ]

wireGrid = [
  
    [850,450],
    [1030,450],
    [1180,450]
]


for x in range(490 , 840, 27):
    for y in range(720 , 900, 27):
        meatLoc.append([x,y])
        

count = 0
for i in range (0,8):
    for net in wireGrid:
        for meat in meatLoc:

            # 每次點擊都會重新取得指定打擊點的 RGB 值
            coord = pyautogui.pixel( meat[0], meat[1] )


            # coord[0] 代表 Red，coord[1] 代表 Green，coord[2] 代表 Blue
            if coord[0] > 170 and coord[2] < 70:
                pyautogui.PAUSE = 0.01
                pyautogui.moveTo( meat[0], meat[1], duration=0.1 )
                sleep(0.1)
                pyautogui.mouseDown()
                pyautogui.moveTo(net[0], net[1], duration=0.1)
                sleep(0.1)
                pyautogui.mouseUp()
                print(f'生肉顏色 {coord}')
                break
            
            
    for j in range(0,2): 
        sleep(4.5)
        for net in wireGrid:
            pyautogui.moveTo(net[0], net[1], duration=0.2)
            sleep(0.1)
            pyautogui.click()



    for net in wireGrid:
        coordmeat = pyautogui.pixel( net[0], net[1] )
        print(f'兩面烤過後第一次檢查 {coordmeat}')

        if coordmeat[0] > 170 and coordmeat[2] < 70:
            pyautogui.moveTo(net[0], net[1], duration=0.2)
            pyautogui.click()
            pass
            
        else: 
            
            pyautogui.moveTo(net[0], net[1], duration=0.2)
            sleep(0.01)
            pyautogui.click()
            coordmeat = pyautogui.pixel( net[0], net[1] )
            print(f'兩面烤過後第二次檢查 {coordmeat}')
            
            if coordmeat[0] > 185 and coordmeat[2] < 173 :
                pyautogui.moveTo(net[0], net[1], duration=0.2)
                pyautogui.click()

            else:

                pyautogui.PAUSE = 0.01
                pyautogui.moveTo(net[0], net[1], duration=0.2)
                sleep(0.01)
                pyautogui.mouseDown()
                pyautogui.moveTo(1310, 850, duration=0.2)
                sleep(0.01)
                pyautogui.mouseUp()
    
    for net in wireGrid:
        coordmeat = pyautogui.pixel( net[0], net[1] )
        print(f'檢查是烤網還是肉 {coordmeat}')
        if coordmeat[0] < 70 and coordmeat[1] < 70 and coordmeat[2] < 70:
            continue
            
        else:
            sleep(2)
            pyautogui.PAUSE = 0.01
            pyautogui.moveTo(net[0], net[1], duration=0.2)
            sleep(0.01)
            pyautogui.mouseDown()
            pyautogui.moveTo(1310, 850, duration=0.2)
            sleep(0.01)
            pyautogui.mouseUp()

            
    
        
```

```python
if __name__ == "__main__":
    while True:
        # 當我們按下 ctrl 鍵時，自動點擊，按下 alt 則結束迴圈
        if keyboard.is_pressed('ctrl'):
            click()
        elif keyboard.is_pressed('alt'):
            break
```
