```python
# 使用 requests 工具
import requests

# 使用 json 工具
import json

import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint
from time import sleep
import random

import re
# 執行 command 的時候用的
import os

```

```python
# 建立儲存圖片、影片的資料夾
folderPath = 'books'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)
```

```python

url = "https://www.gutenberg.org/browse/languages/zh"

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url)

# 指定 lxml 作為解析器
soup = bs(res.text, 'lxml')

link = soup.select('li.pgdbetext > a[href]')
link

```

```python
# 建立 list 來放置列表資訊
booklist = []

#可用正規表達式取得bookid
# regex01 = r'[0-9]{5}'

# 取得 列表 的文字與超連結
for a in soup.select('li.pgdbetext > a[href]'):
    booklist.append({
        'title': a.get_text(),
        'link': 'https://www.gutenberg.org' + a['href'],
        #'book_id':re.search(regex01, a['href'])
        'book_id': a['href'][8:]
    })

pprint(booklist)

# print(booklist[0]['book_id'].group())
# print(booklist[0]['book_id'])
```
